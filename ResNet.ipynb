```python
DIRECTORY =  '/content/drive/MyDrive/dataset'
# a raw string(r'...') is used to treat backslash(\) as a normal character
CATEGORIES = ['mammooty','mohanlal']
```

```python
IMG_SIZE = 224

data = []

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        label = CATEGORIES.index(category)
        img_arr = cv.imread(img_path)
        img_arr = cv.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
#         labelling the data
```

```python
import random
random.shuffle(data)
```

```python
X = []
y = []

for features, labels in data:
    X.append(features)
    y.append(labels)
```

```python
X = np.array(X)
y = np.array(y)
# changed X & y into arrays and stored it in respective variables
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
```

```python
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)
```

```python
model = Model(inputs=base_model.input, outputs=predictions)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
y.shape
```

```python
model.fit(X, y, epochs=5, validation_split=0.2)
```

```python
model.save('resnet.model',save_format = 'h5')
```
