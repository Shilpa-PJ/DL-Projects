```python
import numpy as np
import cv2 as cv
import random
import os
import matplotlib.pyplot as plt
import pickle
import keras
import tensorflow as tf
%matplotlib inline
```

```python
DIRECTORY = r'/content/drive/MyDrive/dataset'
# a raw string(r'...') is used to treat backslash(\) as a normal character
CATEGORIES = ['mammooty','mohanlal']
```

```python
IMG_SIZE = 224

data = []

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        label = CATEGORIES.index(category)
        img_arr = cv.imread(img_path)
        img_arr = cv.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
#         labelling the data
```

```python
import random
random.shuffle(data)
```

```python
X = []
y = []

for features, labels in data:
    X.append(features)
    y.append(labels)
```

```python
X = np.array(X)
y = np.array(y)
# changed X & y into arrays and stored it in respective variables
```

```python
Y=y.reshape(y.shape[0],1,1,)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten,GlobalAveragePooling2D
from tensorflow.keras.models import Model

```

```python
from tensorflow.keras.applications.vgg19 import VGG19

base_model = VGG19(
    include_top=False,
    weights='imagenet',
    input_shape=(224,224,3)
)
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)
```

```python
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
Y.shape
```

```python
model.fit(X, Y, epochs=5, validation_split=0.2)
```

```python
model.save('vgg19.model',save_format = 'h5')
```
