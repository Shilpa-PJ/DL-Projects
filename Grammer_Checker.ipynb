```python
import torch
from transformers import pipeline
```

```python
# Load the pre-trained model for text generation (you can choose other models too)
generator = pipeline("text-generation", model="gpt2")
```

```python
# Input text with possible grammar issues
input_text = "He enjoy playing football."
```

```python
# Use the model to generate a corrected version of the text
corrected_text = generator(input_text, max_length=50, num_return_sequences=1)[0]['generated_text']
```

```python
# Print the corrected text
print("Original Text:", input_text)
print("Corrected Text:", corrected_text)
```
